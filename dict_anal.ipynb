```python
letter_count1 = words5['letter1'].value_counts()
letter_count2 = words5['letter2'].value_counts()
letter_count3 = words5['letter3'].value_counts()
letter_count4 = words5['letter4'].value_counts()
letter_count5 = words5['letter5'].value_counts()
letter_percents1 = round(100 * words5['letter1'].value_counts() / total5, 1)
letter_percents2 = round(100 * words5['letter2'].value_counts() / total5, 1)
letter_percents3 = round(100 * words5['letter3'].value_counts() / total5, 1)
letter_percents4 = round(100 * words5['letter4'].value_counts() / total5, 1)
letter_percents5 = round(100 * words5['letter5'].value_counts() / total5, 1)
letter_percents_df = pd.concat([letter_count1,
                                letter_count2,
                                letter_count3,
                                letter_count4,
                                letter_count5,
                                letter_percents1, 
                                letter_percents2,
                                letter_percents3,
                                letter_percents4,
                                letter_percents5,
                               ],    
                               keys=['count1',
                                     'count2',
                                     'count3',
                                     'count4',
                                     'count5',
                                     'letter1',
                                     'letter2',
                                     'letter3',
                                     'letter4',
                                     'letter5',
                                     ], axis=1).fillna(0)
letter_percents_df['overall'] = round(100 * letter_percents_df.iloc[:, 0:5].sum(axis=1) / (total5*5), 2)
letter_percents_df = letter_percents_df[["letter1", "letter2", "letter3", "letter4", "letter5", "overall"]]
letter_percents_df.sort_values(by='overall', ascending=False)
```

```python
def word_score(word, df=letter_percents_df):
    #check that word is str of correct length
    #check that df has that many columns
    overall_bonus = 2
    arr = list(word)
    score1 = df.loc[arr[0],'letter1'] + df.loc[arr[0],'overall']*overall_bonus
    score2 = df.loc[arr[1],'letter2'] + df.loc[arr[1],'overall']*overall_bonus
    score3 = df.loc[arr[2],'letter3'] + df.loc[arr[2],'overall']*overall_bonus
    score4 = df.loc[arr[3],'letter4'] + df.loc[arr[3],'overall']*overall_bonus
    score5 = df.loc[arr[4],'letter5'] + df.loc[arr[4],'overall']*overall_bonus
    
    ave_score = np.mean([score1, score2, score3, score4, score5])
    ave_score = ave_score * 100 / 32.47
    ave_score = round(ave_score, 2)
    return ave_score
word_score('saaee', letter_percents_df)
```
